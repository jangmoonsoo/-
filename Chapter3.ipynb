```python
x = torch.tensor(data=[2.0,3.0], requires_grad=True)
y=x**2
z=2*y+3

target = torch.tensor([3.0,4.0])
loss = torch.sum(torch.abs(z-target))
loss.backward()

print(x.grad, y.grad, z.grad)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.init as init
```

torch.nn :: 신경망 모델  
torch.optim :: 경사하강법 알고리즘  
torch.nn.init :: 텐서에 초기값을 주는 함수

```python
num_data = 1000
num_epoch = 500

x = init.uniform_(torch.Tensor(num_data,1), -10,10)
noise = init.normal_(torch.FloatTensor(num_data,1), std=1)
y=2*x+3
y_noise = 2*(x+noise)+3
```

torch.Tensor(1000,1) : 1000x1짜리 벡터를 생성하고 그 값을 랜덤하게 채워넣어라  
init.uniform_(A , -10,10) : A값을 -10~10 사이의 유니폼분포에서 샘플링해서 채워라  

```python
model = nn.Linear(1,1)
loss_func = nn.L1Loss()
```

nn.Linear(a,b) : a = 데이터의 변수 갯수, b = 출력할 결과의 변수갯수  

```python
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

optim.SGD : 최적화 알고리즘 중 Stochastic Gradient Descent 방법을 선택  
optim.SGD(model.parameters(), lr=0.01) : 앞에서 정의한 model object에서 설정한 파라미터들을 최적화할 파라미터로 사용, Learning Rate=0.01로 설정

```python
label = y_noise
for i in range(num_epoch) :
    optimizer.zero_grad()
    output = model(x)
    
    loss = loss_func(output, label)
    loss.backward()
    optimizer.step()
    
    if i %10==0:
        print(loss.data)
        
    param_list=list(model.parameters())
    print(param_list[0].item(), param_list[1].item())
```

optimizer.zero_grad() : 에포크가 새로 시작할 때 마다 그레디언트를 0으로 초기화  
output = model(x) : 앞에서 정의한 선형모형인 model 오브젝트에 데이터인 x를 전달, 즉 선형식의 결과를 output에 전달
loss를 L1-loss로 설정 후 backward() 를 통해 선형식의 파라미터인 w,b를 업데이트 (역전파 방법)  
optimizer.step() : 앞에서 정의한 optimizer함수에 인자로 들어가는 model.parameters() 들에 lr을 곱한 뒤 빼서 w,b를 업데이트
